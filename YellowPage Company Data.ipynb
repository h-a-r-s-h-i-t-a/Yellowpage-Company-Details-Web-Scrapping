```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
# Define the URL of the directory website
base_url = "https://www.yellowpages.com/search?search_terms=IT+Services&geo_location_terms=New+York%2C+NY"

# List to store scraped data
companies = []

# Loop through the first 2-3 pages
for page in range(1, 4):
    print(f"Scraping page {page}...")
    html_code = requests.get(base_url + f"&page={page}").text
    soup = BeautifulSoup(html_code, 'lxml')

    company_tag = soup.find_all('div', class_ = 'result')

    for i in company_tag:

        name = i.find('a', class_ = 'business-name').text.strip() if i.find('a', class_ = 'business-name').text.strip() else "No company name found"

        yellowpage_url = f"https://www.yellowpages.com{i.find('a', class_ = 'business-name').get('href')}" if i.find('a', class_='business-name') and i.find('a', class_='business-name').get('href') else "No yellowpage url"

        company_url = i.find('a', class_ = 'track-visit-website')['href'] if i.find('a', class_ = 'track-visit-website') else "No company URL"

        category = ", ".join(link.text for link in i.find('div', class_='categories').find_all('a')) if i.find('div', class_='categories') else 'No Category'

        phone_no = i.find('div', class_='phones phone primary').text.strip() if i.find('div', class_='phones phone primary') else 'Phone no not disclosed'

        street = i.find('div', class_='street-address').text.strip() if i.find('div', class_='street-address') else 'Street not Mentioned'
        
        locality = i.find('div', class_='locality').text.strip() if i.find('div', class_='locality') else 'Locality not mentioned'

        description = i.find('div', class_ = 'snippet').find('p', class_ = 'body').text if i.find('div', class_ = 'snippet') else 'Description not available'
        
        # Combined address
        address = f"{street}, {locality}" if street and locality else street or locality

        
        companies.append({
                "Company Name": name,
                "Website URL": company_url,
                "Yellow Page URL": yellowpage_url,
                "Contact Number": phone_no,
                "Location/Address": address,
                "Industry/Category": category,
                "Company Description": description,
                "Email Address": None  # Email was not exposed of companies
            })
```

```python
Company_details = pd.DataFrame(companies)
```
